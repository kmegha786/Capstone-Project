```python
# Importing libraries

import selenium
import pandas as pd
import time
from bs4 import BeautifulSoup


# Importing selenium webdriver

from selenium import webdriver


# Importing required Exceptions which needs to be handled
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException

# Importing requests

import requests

from selenium.webdriver.common.by import By
```

```python
driver=webdriver.Chrome()
```

```python
driver.get("https://housing.com/rent/house-for-rent-in-mumbai-maharashtra-M2Pskwz0ocdh7q42r5")
```

```python
property_url=[]
start=1
end=11

for page in range(start,end):
    url=driver.find_elements(By.XPATH,'//a[@class="_csbfng _c8dlk8 _frwh2y _ll6yew _1q73uea4 _ks15vq _vv1q9c _sq1l2s"]')
    for i in url:
        property_url.append(i.get_attribute('href'))
    time.sleep(2)  

```

```python
property_url
```

```python
Address=[]
Rent=[]
Baths=[]
Beds=[]
Deposit=[]
Area=[]
Facilities_all=[]
Pets=[]
Washer=[]
AC=[]
Parking=[]
Fireplace=[]
Dishwasher=[]
Floors=[]
Storage=[]
```

```python
for url in property_url:
    driver.get(url)
    time.sleep(2)
    
   


    try:
        address = driver.find_element(By.XPATH,'//div[@class="_h31y44 _csbfng _c8exct _g38jkm _7lc65e"]')
        Address.append(address.text)
    except NoSuchElementException:
        Address.append('-')   
        
    try:
        rent = driver.find_element(By.XPATH,'//span[@class="css-gg3jj9"]')
        Rent.append(rent.text)
    except NoSuchElementException:
        Rent.append(0)
    
    try:
        bedrooms = driver.find_element(By.XPATH,'//div[@class="css-1hidc9c"]')
        Beds.append(bedrooms.text)
    except NoSuchElementException:
        Beds.append('-')
        
    try:
        bathrooms = driver.find_element(By.XPATH,'//*[@id="overviewPoints"]/div/section[2]/table/tbody/tr[6]/td/div')
        Baths.append(bathrooms.text)
    except NoSuchElementException:
        Baths.append(0)
        
    try:
        area = driver.find_element(By.XPATH,'//*[@id="overviewPoints"]/div/section[2]/table/tbody/tr[14]/td')
        Area.append(area.text)
    except NoSuchElementException:
        Area.append(0)
    
    try:
        deposit = driver.find_element(By.XPATH,'//*[@id="overviewPoints"]/div/section[2]/table/tbody/tr[2]/td/div')
        Deposit.append(deposit.text)
    except NoSuchElementException:
        Deposit.append(0)
        
    # Facilities Data
    try:

        facilities  = WebDriverWait(driver,3).until(EC.presence_of_all_elements_located((By.XPATH,'//div[@class="_be1g80 _9s1txw item-container"]')))

        Facilities_all = []
        for i in facilities:
            facilities_all.append(i.text)
    except:
        facilities_all = ['null','null']


    if 'Pets_Allowed' in Facilities_all:
        Pets.append('Yes')
    else :
        Pets.append('No')

    if 'Washer/Dryer' in Facilities_all:
        Washer.append('Yes')
    else :
        Washer.append('No')

    if 'Ac' in Facilities_all:
        AC.append('Yes')
    else :
        AC.append('No')

    if 'parking' in Facilities_all:
        Parking.append('Yes')
    else :
        Parking.append('No')

    if 'fireplace' in Facilities_all:
        Fireplace.append('Yes')
    else :
        Fireplace.append('No')

    if 'dishwasher' in Facilities_all:
        Dishwasher.append('Yes')
    else :
        Dishwasher.append('No')

    if 'Hardwood_Floors' in Facilities_all:
        Floors.append('Yes')
    else :
        Floors.append('No')
        
    if 'Roofdeck_storage' in Facilities_all:
        Storage.append('Yes')
    else :
        Storage.append('No')
```

```python


d={'Address':Address,'Rent':Rent,'Deposit':Deposit,'Beds':Beds,'Baths':Baths,'Area':Area,'Pets_Allowed':Pets,'Washer/Dryer':Washer,'AC':AC,'Parking':Parking,'Fireplace':Fireplace,'Dishwasher':Dishwasher,'Hardwood_Floors':Floors,'Roofdeck_Storage':Storage}
```
